For the purpose of raster data analysis GeoBrix provides functions for distributed GDAL data ingestion, focused on GTiff ('tif'). After loading the underlying files directly, we also add new columns from the following via function `finalize_tiled_band_tbl`:

* `rst_memsize` for 'size'

* `rst_boundingbox` to generate tile bounding boxes
* `rst_srid` to pull out 'srid' from the raster metadata

__Notes:__

1. Since this involves a bit of computation, we persist out as Delta Table to then use multiple times below.
1. Operation can take a few minutes using the configuration described at top of notebook (more if doing all of Alaska).
1. We also OPTIMIZE (table layout) based on item_id which will help future ops.

> **Serverless note:** the `gtiff` reader is scoped **per band** (`tifs_for(band)`) and splits each ~240 MB scene into ~32 MB sub-tiles (`sizeInMB`) so the `rst_*` UDFs stay under Serverless’s ~1 GB per-task cap; it stages each scene to worker-local disk before windowing (UC Volume FUSE can’t seek). `finalize_tiled_band_tbl` repartitions to one tile per task (Serverless can’t raise shuffle partitions).

__Read the downloaded tifs.__

> This is a reader for all the tifs, but we inner join on various bands of interest to put them into separate tables.

```python
ls -lh $EO_DIR
```

```python
def tifs_for(band):
  # Per-band read: load ONLY this band’s subdir. Reading all 4 bands’ full-res
  # COGs through the reader at once exceeds Serverless worker memory; one band’s
  # files (joined per band) stays safe. The reader splits each scene into ~32 MB
  # sub-tiles (sizeInMB) via in-memory windowing (UC Volume FUSE can’t seek).
  return (
    spark
      .read
      .format("gtiff_gbx")
        .option("sizeInMB", "32")
        .option("filterRegex", r".*[.]tif$")
      .load(f"{EO_DIR}/{band}")
    .withColumn("out_file_path", F.replace("source", F.lit("dbfs:/"), F.lit("/")))
    .drop("source")
  )
```

```python
b02_df = b02_df.join(tifs_for("B02"), on=['out_file_path'])
```

## TIF Summary Information

`rst_summary` returns a JSON string with per-tile metadata including dimensions, coordinate system, band info, and nodata values. See [RasterX Functions](https://databrickslabs.github.io/geobrix/docs/api/raster-functions).

```python
summary_df = (
  df_tile
    .withColumn("summary_tile", rx.rst_summary("tile"))
    .selectExpr("summary_tile", "* except(summary_tile)")
)
summary_df.limit(1).show(vertical=True) # <- show + limiting for ipynb only
```

> [2b] Example parsing out band information from summary, in this example we count 'band.noDataValue' (which we previously standardized). __Note: there are other helper functions as well for metadata not shown, e.g. `rst_bandmetadata`.__ We haven't yet stacked bands (will do in notebook 04), so dealing with single band TIFFs up to now.

```python
# we use an exemplar to further parse structured metadata from the summary, in this case for bands. 
# - then we group by the numeric value for `band.noDataValue` and see if all is consistent
# - it is because we previously standardized it
example_band_obj = """[{"band":1,"block":[512,512],"type":"UInt16","colorInterpretation":"Gray","noDataValue":0.0,"overviews":[{"size":[5490,5490]},{"size":[2745,2745]},{"size":[1373,1373]},{"size":[687,687]},{"size":[344,344]}],"metadata":{}}]"""

(
    summary_df   
        .selectExpr("summary_tile:['bands'] as band_obj")
        .select(F.from_json("band_obj", F.schema_of_json(example_band_obj)).alias("bands"))
        .select(F.explode("bands").alias("band"))    
        .groupBy("band.noDataValue")
            .count()
).display()
```

## K-Rings

> Once your rasters have been gridded, you can take advantage of calls such as K-Rings.

_We are just doing a kring around a H3 resolution 5 area in this example. Since our data is @ resolution 7, we can use the `h3_toparent` function available in product._

```python
h3_df = spark.table("band_b02_h3")
```

```python
kring_cellid = (
  h3_df
    .groupBy("cellid")
      .count()
    .orderBy(F.desc("count"))
    .selectExpr("h3_toparent(cellid, 5)")
).first()[0]

kring_cellid
```

_Also, we are choosing to go out 3 rings from our center cell of interest in this example, using built-in product [h3_kring](https://docs.databricks.com/en/sql/language-manual/functions/h3_kring.html) for just the h3 cells around the provide one(s)._

`rst_merge_agg` merges all tiles whose H3 cell falls within the k-ring neighborhood into a single raster — useful for building spatial context around a cell of interest. Note: `h3_kring` and `h3_toparent` are Databricks built-in H3 functions, not GeoBrix.

```python
kring_df = (
  h3_df
    .filter(DBF.h3_toparent("cellid", 5) == kring_cellid)
    .withColumn("kring", F.expr("explode(h3_kring(cellid, 3))"))
    .groupBy("kring", "date")
      .agg(rx.rst_merge_agg("tile").alias("tile"))
)
print(f"count? {kring_df.count():,}")
kring_df.limit(1).show(vertical=True) # <- show + limiting for ipynb only
```

## Raster to Timeseries projection

> __With the power of expression that rasterio provides + the power of distribution that GeoBrix provides__, we can easily convert rasters to numerical values with arbitrary mathematical complexity. Since all of our imagery is timestamped, our raster to number projection in effect is creating time series bound to H3 cells.

```python
def rio_mean_band(dataset):
  try:
    return dataset.statistics(bidx = 1).mean
  except:
    return 0.0

measure_df = kring_df.withColumn(
  "rio_mean", library.rasterio_lambda("tile.raster", lambda dataset: rio_mean_band(dataset) )
)
```

```python
daily_measure_df = (
  measure_df
    .selectExpr(
      "kring as cellid",
      "date",
      "rio_mean",
      "tile.raster"
    )
    .where("rio_mean > 0")
    .groupBy("cellid", "date")
    .agg(
      F.avg("rio_mean").alias("measure"),
      F.first("raster").alias("raster")
    )
)
daily_measure_df.limit(1).show(vertical=True) # <- show + limiting for ipynb only
```

_At this point our data effectively became timeseries data and can be modeled as virtual IOT  devices that are fixed in spatial context. We can easily visualise data for individual dates in a spatial contex by leveraging our H3 locations._

```python
df_filter_date = daily_measure_df.where(f"date == '{filter_date}'")
```
